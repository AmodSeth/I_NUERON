```python
from sklearn.datasets import make_classification

```

```python
#n_samples=1000 n_cluster_per_class=1 to create a grouping of each feature
## X = independent features
## y = dependet features

X,y=make_classification(n_samples=1000, n_redundant=0,n_clusters_per_class=1,weights=[0.90],n_features=2,random_state=1)
```

```python
import pandas as pd
D1 = pd.DataFrame(X,columns=['f1','f2'])
D2 = pd.DataFrame(y,columns=['target'])


final_df=pd.concat([D1,D2],axis=1)
```

```python
final_df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE

### TRANSFORMING THE DATASET
#isntance crete
oversample=SMOTE()


### datapoints and target it takes as parameters
X,y=oversample.fit_resample(final_df[['f1','f2']],final_df['target'])

```

```python
X.shape,y.shape
```

```python
d1=pd.DataFrame(X,columns=['f1','f2'])
d2=pd.DataFrame(y,columns=["target"])

## axis=1 mneans column wise add
oversamples=pd.concat([d1,d2],axis=1)
```

```python
import matplotlib.pyplot as plt
plt.scatter(oversamples['f1'],oversamples['f2'],c=oversamples['target'])
```
